## Initial AOV evaluation

1b. The metric I would report for this dataset would be the weighted average to properly consider the number of sneakers purchased in each order. 

#### 1c. Calculating the Weighted Average

```python
total_order_amount = df1['order_amount'].sum()
total_shoes_ordered = df1['total_items'].sum()
weighted_average = total_order_amount/total_shoes_ordered
```

### 2a. How many orders were shipped by Speedy Express in total?
### ANSWER: 54

#### SQL QUERY:

```python
SELECT
Count (DISTINCT OrderID)
FROM Orders
WHERE ShipperID=1
```

NOTES: The distinct function is called to handle any orderID duplicates that might be in the data.

### 2b. What is the last name of the employee with the most orders?
### ANSWER: Peacock

#### SQL QUERY:

```python
SELECT lastname, MAX (ordercount) FROM 
    (
        SELECT o.employeeid, lastname, count(orderid) ordercount
         FROM orders o
            JOIN employees em ON o.employeeid = em.employeeid
        GROUP BY o.employeeid
    )
```

### 2c. What product was ordered the most by customers in Germany?
### ANSWER: Gorgonzola Telino

#### SQL QUERY:

```python
SELECT productid, productname, MAX(ProductCount) FROM 
    (
          SELECT o.orderid, c.customerid, orderdetailid, od.productid, count(od.productid) ProductCount
          FROM Orders o
          JOIN customers c on o.customerid = c.customerid
          JOIN orderdetails od on o.orderid = od.orderid
          WHERE country = 'Germany'
          GROUP BY productid
     )
JOIN products USING (productid)
```
